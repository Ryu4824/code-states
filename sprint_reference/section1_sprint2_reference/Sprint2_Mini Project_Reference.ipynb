- 처음에 1번 문을 선택했다고 가정합니다. 진행자가 남은 두 개의 문 중 염소가 있는 문 하나를 공개한 상태에서 기존에 선택했던 문(1번문)에 자동차가 있을 확률을 구하세요. 

- 처음 선택을 유지하는 것이 유리한지, 아니면 선택을 바꾸는 것이 유리한지 생각해보세요.

```
Hypothesis: 1번 문 뒤에 자동차가 있음
Evidence: 진행자가 염소가 있는 문을 열어줌

우리가 알고자 하는 것: P(H|E) -> 진행자가 염소가 있는 문을 열었을 때, 우리가 먼저 선택했던 문 안에 자동차가 있을 확률
```


#### **1-1. 진행자가 남은 두 개의 문 중 염소가 있는 문 하나를 공개한 상태에서 기존에 선택했던 문에 자동차가 있을 확률을 구하세요.**
- 처음에 1번 문을 선택했다고 가정합니다.

````python
Hypothesis = 1/3
not_Hypothesis = 2/3

Evidence_Hypothesis = 1
Evidence_not_Hypothesis = 1

Hypothesis_Evidence = (Evidence_Hypothesis * Hypothesis) / ((Evidence_Hypothesis * Hypothesis) + (Evidence_not_Hypothesis * not_Hypothesis))

Hypothesis_Evidence
````

### `Topic 2` : 베이지안 정리 적용

#### **2-1. 고객이 리조트 호텔을 예약했을 때, 예약을 취소할 확률은 어떻게 변하나요?**
> Bayesian Theorem의 공식은 아래와 같습니다. <br>
<br>
> $P(cancel|resort) = P(cancel \cap resort) / P(resort) = P(resort|cancel)P(cancel) / P(resort)$
<br>

- 예약을 취소하는 사전 확률($P(cancel)$)은 **이유 불충분의 원리**에 따라 `0.5`로 설정합니다.
- 데이터셋을 바탕으로, 아래의 확률을 도출해보세요.
  - 예매를 취소했는데, Resort Hotel을 예약한 경우 ($P(resort|cancel)$) : `resort_yes_cancel`
  - 예매를 취소하지 않았는데, Resort Hotel을 예약한 경우 ($P(resort|not \; cancel)$) : `resort_no_cancel`
- 상기한 경우들을 바탕으로, 베이지안 정리를 통해 사전 확률을 업데이트하는 함수를 완성해보세요.
- 다음 `bayesian_hotel` 함수를 완성하여 사후확률을 구하는 것이 어렵다면, warm-up 영상에서 나온 사각형 방식으로 사후확률을 계산해보세요.



````python
# 예매를 취소했는데, Resort Hotel을 예약한 경우
# 예매를 취소하지 않았는데, Resort Hotel을 예약한 경우
# Resort Hotel을 예약했을 때, 예매를 취소할 확률 (posterior)

def bayesian_hotel(prior, resort_yes_cancel, resort_no_cancel):
  # 사전 확률
  cancel = prior
  no_cancel = 1 - prior
  
  # evidence
  # 예매를 취소했는데, Resort Hotel을 예약한 경우
  likelihood = resort_yes_cancel
  # 예매를 취소하지 않았는데, Resort Hotel을 예약한 경우 = resort_no_cancel  

  # 사후 확률
  posterior = (likelihood * cancel) / ((likelihood * cancel) + (resort_no_cancel * no_cancel))
  return posterior
````

````python
# 고객이 예약을 취소하는 경우
cancel_cond = (hotel["is_canceled"]==1)

# P(resort_hotel|cancel)
# 예매를 취소했는데, Resort Hotel을 예약한 경우
resort_yes_cancel = hotel[cancel_cond==True]["hotel"].value_counts(normalize=True)["Resort Hotel"]

# P(resort_hotel|no_cancel)
# 예매를 취소하지 않았는데, Resort Hotel을 예약한 경우
resort_no_cancel = hotel[cancel_cond==False]["hotel"].value_counts(normalize=True)["Resort Hotel"]


bayesian_hotel(0.5, resort_yes_cancel, resort_no_cancel)
````

💡 **해설 Tip)**

> **큰 수의 법칙 (Law of Large Numbers)**
: Sample Size 커질수록, Sample의 통계치는 모집단의 모수에 가까워 진다.

> **중심극한정리 (Central Limit Theorem)**
: 모집단의 분포와 상관없이, 임의의 분포에서 추출된 표본 평균의 분포는 정규분포를 이룬다.

> **부트스트래핑 (Bootstrapping)**
: 샘플을 모집단이라 가정한 후, 여기에서 다시 샘플을 추출하여 샘플들의 평균의 분포를 확인 (Resampling)
- 중복 추출 허용
- 모수를 더 잘 이해하기 위해 추가적인 데이터 필요 없이, 현재 갖고 있는 데이터로 추정 가능

> **신뢰 구간 (Confidence Interval)**
: 모수가 존재할 가능성이 높은 구간
- 모수의 평균을 알 수 없기 때문에 모수를 특정 값으로 추정하게 되면 불확실성을 초래한다.
- 모수가 존재할 가능성이 높은 구간 (신뢰구간)을 확률 (신뢰수준)과 함께 제공 <br>
    ⇒ 불확실성 줄이고, 모수의 신뢰성 가늠

````python
# lead_time 컬럼 분포 및 평균 비교
lead_time_mean = hotel["lead_time"].mean()

plt.figure(figsize=(8,5))
lead_time_value = hotel["lead_time"].values
sns.distplot(lead_time_value)
# lead_time 평균을 초록 선으로 표시
plt.axvline(lead_time_mean, c="green", lw=2, label="Mean of Lead Time")
plt.title("lead_time 컬럼 분포", fontsize=13, fontweight="bold")
plt.legend()
plt.show()
````

````python
np.random.seed(42)

# 3, 10, 30개의 데이터 랜덤하게 선택
sample1 = np.random.choice(lead_time_value, 3)
sample2 = np.random.choice(lead_time_value, 10)
sample3 = np.random.choice(lead_time_value, 30)

# 랜덤하게 선택한 데이터들의 평균값 구한다
sample_mean_3 = sample1.mean()
sample_mean_10 = sample2.mean()
sample_mean_30 = sample3.mean()

# 랜덤하게 선택한 데이터들의 평균값을 모집단의 평균과 비교
plt.figure(figsize=(8,5))
plt.hist(lead_time_value,alpha=0.5)
plt.axvline(lead_time_mean, c='green', lw=2, label='Population Mean')
plt.axvline(sample_mean_3, c='blue', lw=2, label='Mean of Sample Size 3')
plt.axvline(sample_mean_10, c='yellow', lw=2, label = 'Mean of Sample Size 10')
plt.axvline(sample_mean_30, c='orange', lw=2, label = 'Mean of Sample Size 30')
plt.title("랜덤하게 선택한 데이터들의 평균값을 모집단의 평균과 비교", fontsize=13, fontweight="bold")
plt.legend();
````

````python
# 중심극한정리를 통해 정규분포 모양 갖는 것 확인

np.random.seed(42)

# lead_time 컬럼에서 샘플 추출하여 평균값 계산한 list
lt_means_size_30 = []

# arrival_date_day_of_month 컬럼에서 샘플 추출하여 평균값 계산한 list
add_means_size_30 = []

# arrival_date_month 컬럼에서 샘플 추출하여 평균값 계산한 list
adm_means_size_30 = []

for _ in range(10000):
  sample_lt = np.random.choice(hotel["lead_time"], 30) # 간단하게 살펴보기 위해 sample size 30으로 지정
  sample_add = np.random.choice(hotel["arrival_date_day_of_month"], 30)
  sample_adm = np.random.choice(hotel["arrival_date_month"], 30)
  
  # 각 평균값들을 빈 list에 삽입
  lt_means_size_30.append(sample_lt.mean())
  add_means_size_30.append(sample_add.mean())
  adm_means_size_30.append(sample_adm.mean())

fig, axes = plt.subplots(1, 3, figsize=(16, 8))
fig.suptitle("Central Limit Theorem (Sample Size = 30)", fontsize=13, fontweight="bold")

sns.distplot(lt_means_size_30, ax=axes[0])
axes[0].set_title("lead_time 컬럼 평균 표본 분포")

sns.distplot(add_means_size_30, ax=axes[1])
axes[1].set_title("arrival_date_day_of_month 컬럼 평균 표본 분포")

sns.distplot(adm_means_size_30, ax=axes[2])
axes[2].set_title("arrival_date_month 컬럼 평균 표본 분포")
plt.show()
````

  - 신뢰 수준 90%, 95%, 99%를 갖는 신뢰구간을 그려보고 구간의 크기가 어떻게 변하는지 확인해보세요.
  - 신뢰구간이 넓어지거나 좁아졌을 때의 장단점에 대해 생각해보세요. (e.g. 어떤 경우에는 신뢰구간을 넓게 설정하는 것이 좋을까요?)

````python
# lead_time 컬럼에서 샘플 추출하여 평균값 계산한 array
lt_means_size_30 = np.array(lt_means_size_30)

# arrival_date_day_of_month 컬럼에서 샘플 추출하여 평균값 계산한 array
add_means_size_30 = np.array(add_means_size_30)

# arrival_date_month 컬럼에서 샘플 추출하여 평균값 계산한 array
adm_means_size_30 = np.array(adm_means_size_30)

# 각 array들의 평균값
lt_mean = lt_means_size_30.mean()
add_mean = add_means_size_30.mean()
adm_mean = adm_means_size_30.mean()

# 각 array의 90% 신뢰구간
lt_lower_90, lt_upper_90 = np.percentile(lt_means_size_30, 5), np.percentile(lt_means_size_30, 95) 
add_lower_90, add_upper_90 = np.percentile(add_means_size_30, 5), np.percentile(add_means_size_30, 95) 
adm_lower_90, adm_upper_90 = np.percentile(adm_means_size_30, 5), np.percentile(adm_means_size_30, 95) 

# 각 array의 95% 신뢰구간
lt_lower_95, lt_upper_95 = np.percentile(lt_means_size_30, 2.5), np.percentile(lt_means_size_30, 97.5) 
add_lower_95, add_upper_95 = np.percentile(add_means_size_30, 2.5), np.percentile(add_means_size_30, 97.5) 
adm_lower_95, adm_upper_95 = np.percentile(adm_means_size_30, 2.5), np.percentile(adm_means_size_30, 97.5) 

# 각 array의 99% 신뢰구간
lt_lower_99, lt_upper_99 = np.percentile(lt_means_size_30, 0.5), np.percentile(lt_means_size_30, 99.5) 
add_lower_99, add_upper_99 = np.percentile(add_means_size_30, 0.5), np.percentile(add_means_size_30, 99.5) 
adm_lower_99, adm_upper_99 = np.percentile(adm_means_size_30, 0.5), np.percentile(adm_means_size_30, 99.5) 


fig, axes = plt.subplots(1, 3, figsize=(16, 8))
fig.suptitle("Confidence Interval (90%, 95%, 99%)", fontsize=13, fontweight="bold")

sns.distplot(lt_means_size_30, ax=axes[0])
axes[0].set_title("lead_time 컬럼 평균 표본 분포")
axes[0].axvline(lt_lower_90, c='dimgray', label='90% Confidence Interval')
axes[0].axvline(lt_upper_90, c='dimgray')
axes[0].axvline(lt_lower_95, c='blue', label='95% Confidence Interval')
axes[0].axvline(lt_upper_95, c='blue')
axes[0].axvline(lt_lower_99, c='dodgerblue', label='99% Confidence Interval')
axes[0].axvline(lt_upper_99, c='dodgerblue')
axes[0].axvline(lt_mean, c='green', label='Mean')
axes[0].legend()

sns.distplot(add_means_size_30, ax=axes[1])
axes[1].set_title("arrival_date_day_of_month 컬럼 평균 표본 분포")
axes[1].axvline(add_lower_90, c='dimgray', label='90% Confidence Interval')
axes[1].axvline(add_upper_90, c='dimgray')
axes[1].axvline(add_lower_95, c='blue', label='95% Confidence Interval')
axes[1].axvline(add_upper_95, c='blue')
axes[1].axvline(add_lower_99, c='dodgerblue', label='99% Confidence Interval')
axes[1].axvline(add_upper_99, c='dodgerblue')
axes[1].axvline(add_mean, c='green', label='Mean')
axes[1].legend()

sns.distplot(adm_means_size_30, ax=axes[2])
axes[2].set_title("arrival_date_month 컬럼 평균 표본 분포")
axes[2].axvline(adm_lower_90, c='dimgray', label='90% Confidence Interval')
axes[2].axvline(adm_upper_90, c='dimgray')
axes[2].axvline(adm_lower_95, c='blue', label='95% Confidence Interval')
axes[2].axvline(adm_upper_95, c='blue')
axes[2].axvline(adm_lower_99, c='dodgerblue', label='99% Confidence Interval')
axes[2].axvline(adm_upper_99, c='dodgerblue')
axes[2].axvline(adm_mean, c='green', label='Mean')
axes[2].legend()

plt.show()
````

  - $H_0$ : 정규성을 만족한다.
  - $H_1$ : 정규성을 만족하지 않는다.

- 일반적으로 sample size가 30 이상인 경우, 중심극한정리에 의해 표본 평균 분포가 정규분포를 따른다고 가정합니다.

- 그 외 정규성을 검정하는 방법인 `Shaprio-Wilks test`, `Kolmogorove-Smirnov test`에 대해 알아보세요.
  - `scipy`
 공식 문서 : [Shapiro–Wilk test](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.shapiro.html), [Kolmogorove-Smirnov test](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ks_2samp.html)



#### **2-3. 등분산 검정**
- 등분산은 독립적인 sample 들의 분포가 비교 가능하고 유사하다는 것을 의미합니다.
  - 등분산 : 2개의 모집단에서 추출된 각 sample의 분산이 같다
- 등분산 검정 또한 `t-test` 사용하기 앞서 확인해봐야 하는 요소입니다.
  - $H_0$ : 등분산이다.
  - $H_1$ : 등분산이 아니다. <br>

<img src="https://i.imgur.com/yfWCtmy.png">

- `Levene's test`, `Bartlett's test`에 대해 알아보세요.
  - `scipy` 공식 문서 : [Levene's test](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.levene.html), [Bartlett's test](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.bartlett.html)


💡 **해설 Tip)** <br>
> `Levene's test`를 사용하여 등분산성을 확인해봅시다.
- 아래의 코드에 따라 `p-value < 0.05`이기 때문에, 등분산성을 만족하지 않습니다.


````python
# 예약 취소 여부
cancel_cond = (hotel["is_canceled"]==1)
# 취소한 예약들의 adr
cancel_adr = hotel[cancel_cond==True]["adr"]
# 취소하지 않은 예약들의 adr
no_cancel_adr = hotel[cancel_cond==False]["adr"]

# Levene's test
stats.levene(cancel_adr, no_cancel_adr)
````

#### **2-4. Two-Sample T-test**
- 위에서 진행한 검정들을 바탕으로, 두 샘플의 평균값을 비교하는 검정입니다.
  - $H_0$ : $\mu_{cancel} = \mu_{no\ cancel}$
  - $H_1$ : $\mu_{cancel} ≠ \mu_{no\ cancel}$
- `ttest_ind (독립 표본 t-test)`에 대해 알아보세요.
  - `scipy` 공식 문서 : [ttest_ind](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html)
  - 공식 문서의 `parameter`들이 의미하는 것에 주의해서 살펴보세요.
    - 등분산성에 대한 파라미터가 무엇인지 확인해보고, 이를 어떻게 설정하면 좋을지 생각해보세요.
- `adr` 컬럼 대신 다른 컬럼들로도 위의 과정을 통해 가설을 수립하고 통계적으로 검정해보세요! (optional)

````python
stats.ttest_ind(cancel_adr, no_cancel_adr, equal_var=False, alternative="two-sided")
````

💡 **해설 Tip)** <br>
> 등분산성을 만족하지 않기 때문에, `equal_var=False`을 지정합니다.<br>
양측 검정을 진행하기 때문에, `alternative="two-sided"`을 지정합니다 (default option)
- 아래의 코드에 따라 `p-value < 0.05`이므로, 귀무가설을 기각합니다.
  - 신뢰수준 95%하에, 취소된 예약과 취소되지 않은 예약의 평균 `adr`은 같지 않다는 **대립가설은 통계적으로 유의합니다**.

#### **3-2. 정규성 검정**
- `z-test`를 활용하기 위해서는, 샘플의 확률분포가 정규분포를 따라야 합니다.
- `이항분포`의 경우, 정규분포를 따르는지 여부를 `reference pdf 파일의 30 page`를 통해 확인하세요. 
  - [이항분포 reference](https://viewpds.jihak.co.kr/tsoldb/%EA%B5%90%EA%B3%BC%EC%84%9C%EB%B3%84/%EA%B3%A0_%EC%A0%81%EB%B6%84%EA%B3%BC%20%ED%86%B5%EA%B3%84(%EC%9D%B4%EA%B0%95%EC%84%AD)/4-1/%EC%A7%80%ED%95%99_%EC%A0%81%EB%B6%84%EA%B3%BC%20%ED%86%B5%EA%B3%84_%EA%B3%A0_%EC%9D%B4_4_1_%ED%99%95%EB%A5%A0%EB%B6%84%ED%8F%AC_%EA%B5%90%EA%B3%BC%EC%84%9C%20%EB%B3%B8%EB%AC%B8.pdf)
- 아래의 코드를 실행하여, 이항분포가 어떤 모양을 갖는지 살펴보세요.


````python
# 확률변수 X가 이항분포 B(n, p)를 따를 때, n이 충분히 크다면 X는 정규분포 N(np, npq)를 따른다 (단, q=1-p)
from math import factorial

# Probability density of the binomial distribution
# p의 확률을 갖는 n회의 시행에서 k번 성공할 확률을 반환하는 bin_dist 함수 생성
def bin_dist(k, n, p):
    nck = factorial(n) / (factorial(k) * factorial(n - k))
    pd = nck * p**k * (1-p)**(n-k)
    return pd

x = np.arange(101)

# 100회 시행했을 때, 1이 나올 확률 
# (e.g. 앞면이 나올 확률이 ab_hotel.is_canceled.mean()인 동전을 100번 던졌을 때, 앞면이 나올 확률)
pd1 = np.array([bin_dist(k, (101-1), ab_hotel.is_canceled.mean()) for k in range(101)])
# y축 범위 지정
plt.ylim(0, 0.1)

# 그래프상의 x, y위치에 문자열 나타내는 plt.text
plt.text(80, 0.095, f'n, p = {101-1}, {round(ab_hotel.is_canceled.mean(), 2)}')
plt.bar(x, pd1, color='lightcoral')
plt.title("예약 취소율 100건의 이항 분포", fontsize=13, fontweight="bold")
plt.show()
````

#### **3-3. `z-test`를 사용할 수 있는 조건**

  - `모집단의 표준편차를 알 수 없지만 표본 크기가 30보다 큰 경우` : **분포가 정규분포에 가깝기 때문에 z-test를 사용 가능**합니다.

4. **Sample Size는 같은 크기**여야 합니다.
5. 각각의 **데이터는 모집단에서 같은 확률로 랜덤하게 선택**되어야 합니다.

#### **3-4. `z-test`**
- `z-test`를 활용하여, 가설을 검정해봅시다.
- statsmodel 공식 문서 : [statsmodels.stats.proportion.proportions_ztest](https://www.statsmodels.org/stable/generated/statsmodels.stats.proportion.proportions_ztest.html)
  - 공식 문서의 `parameter`들이 의미하는 것에 주의해서 살펴보세요.


````python
import statsmodels.api as sm

# 대조군
num_ctrl = ab_hotel.query('button == "top"').shape[0]
num_exp = ab_hotel.query('button == "bottom"').shape[0]

# 대조군, 실험군 그룹 내 retenton이 True인 유저수
top_cancel = ab_hotel.query('button == "top"').is_canceled.sum()
bottom_cancel = ab_hotel.query('button == "bottom"').is_canceled.sum()

# Z-test
z_score, p_value = sm.stats.proportions_ztest([top_cancel, bottom_cancel], 
                                              [num_ctrl, num_exp], 
                                              alternative= 'larger')

print(f'p-value (Z-test): {p_value}')
````
