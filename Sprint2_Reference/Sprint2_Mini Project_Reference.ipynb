### `Topic 2` : ë² ì´ì§€ì•ˆ ì •ë¦¬ ì ìš©

#### **2-1. ê³ ê°ì´ ë¦¬ì¡°íŠ¸ í˜¸í…”ì„ ì˜ˆì•½í–ˆì„ ë•Œ, ì˜ˆì•½ì„ ì·¨ì†Œí•  í™•ë¥ ì€ ì–´ë–»ê²Œ ë³€í•˜ë‚˜ìš”?**
> Bayesian Theoremì˜ ê³µì‹ì€ ì•„ë˜ì™€ ê°™ìŠµë‹ˆë‹¤. <br>
<br>
> $P(cancel|resort) = P(cancel \cap resort) / P(resort) = P(resort|cancel)P(cancel) / P(resort)$
<br>

- ì˜ˆì•½ì„ ì·¨ì†Œí•˜ëŠ” ì‚¬ì „ í™•ë¥ ($P(cancel)$)ì€ **ì´ìœ  ë¶ˆì¶©ë¶„ì˜ ì›ë¦¬**ì— ë”°ë¼ `0.5`ë¡œ ì„¤ì •í•©ë‹ˆë‹¤.
- ë°ì´í„°ì…‹ì„ ë°”íƒ•ìœ¼ë¡œ, ì•„ë˜ì˜ í™•ë¥ ì„ ë„ì¶œí•´ë³´ì„¸ìš”.
  - ì˜ˆë§¤ë¥¼ ì·¨ì†Œí–ˆëŠ”ë°, Resort Hotelì„ ì˜ˆì•½í•œ ê²½ìš° ($P(resort|cancel)$) : `resort_yes_cancel`
  - ì˜ˆë§¤ë¥¼ ì·¨ì†Œí•˜ì§€ ì•Šì•˜ëŠ”ë°, Resort Hotelì„ ì˜ˆì•½í•œ ê²½ìš° ($P(resort|not \; cancel)$) : `resort_no_cancel`
- ìƒê¸°í•œ ê²½ìš°ë“¤ì„ ë°”íƒ•ìœ¼ë¡œ, ë² ì´ì§€ì•ˆ ì •ë¦¬ë¥¼ í†µí•´ ì‚¬ì „ í™•ë¥ ì„ ì—…ë°ì´íŠ¸í•˜ëŠ” í•¨ìˆ˜ë¥¼ ì™„ì„±í•´ë³´ì„¸ìš”.
- ë‹¤ìŒ `bayesian_hotel` í•¨ìˆ˜ë¥¼ ì™„ì„±í•˜ì—¬ ì‚¬í›„í™•ë¥ ì„ êµ¬í•˜ëŠ” ê²ƒì´ ì–´ë µë‹¤ë©´, warm-up ì˜ìƒì—ì„œ ë‚˜ì˜¨ ì‚¬ê°í˜• ë°©ì‹ìœ¼ë¡œ ì‚¬í›„í™•ë¥ ì„ ê³„ì‚°í•´ë³´ì„¸ìš”.



````python
# ì˜ˆë§¤ë¥¼ ì·¨ì†Œí–ˆëŠ”ë°, Resort Hotelì„ ì˜ˆì•½í•œ ê²½ìš°
# ì˜ˆë§¤ë¥¼ ì·¨ì†Œí•˜ì§€ ì•Šì•˜ëŠ”ë°, Resort Hotelì„ ì˜ˆì•½í•œ ê²½ìš°
# Resort Hotelì„ ì˜ˆì•½í–ˆì„ ë•Œ, ì˜ˆë§¤ë¥¼ ì·¨ì†Œí•  í™•ë¥  (posterior)

def bayesian_hotel(prior, resort_yes_cancel, resort_no_cancel):
  # ì‚¬ì „ í™•ë¥ 
  cancel = prior
  no_cancel = 1 - prior
  
  # evidence
  # ì˜ˆë§¤ë¥¼ ì·¨ì†Œí–ˆëŠ”ë°, Resort Hotelì„ ì˜ˆì•½í•œ ê²½ìš°
  likelihood = resort_yes_cancel
  # ì˜ˆë§¤ë¥¼ ì·¨ì†Œí•˜ì§€ ì•Šì•˜ëŠ”ë°, Resort Hotelì„ ì˜ˆì•½í•œ ê²½ìš° = resort_no_cancel  

  # ì‚¬í›„ í™•ë¥ 
  posterior = (likelihood * cancel) / ((likelihood * cancel) + (resort_no_cancel * no_cancel))
  return posterior
````

````python
# ê³ ê°ì´ ì˜ˆì•½ì„ ì·¨ì†Œí•˜ëŠ” ê²½ìš°
cancel_cond = (hotel["is_canceled"]==1)

# P(resort_hotel|cancel)
# ì˜ˆë§¤ë¥¼ ì·¨ì†Œí–ˆëŠ”ë°, Resort Hotelì„ ì˜ˆì•½í•œ ê²½ìš°
resort_yes_cancel = hotel[cancel_cond==True]["hotel"].value_counts(normalize=True)["Resort Hotel"]

# P(resort_hotel|no_cancel)
# ì˜ˆë§¤ë¥¼ ì·¨ì†Œí•˜ì§€ ì•Šì•˜ëŠ”ë°, Resort Hotelì„ ì˜ˆì•½í•œ ê²½ìš°
resort_no_cancel = hotel[cancel_cond==False]["hotel"].value_counts(normalize=True)["Resort Hotel"]


bayesian_hotel(0.5, resort_yes_cancel, resort_no_cancel)
````

- ìœ„ì—ì„œ ì§„í–‰í•œ ê²€ì •ë“¤ì„ ë°”íƒ•ìœ¼ë¡œ, ë‘ ìƒ˜í”Œì˜ í‰ê· ê°’ì„ ë¹„êµí•˜ëŠ” ê²€ì •ì…ë‹ˆë‹¤.
  - $H_0$ : $\mu_{cancel} = \mu_{no\ cancel}$
  - $H_1$ : $\mu_{cancel} â‰  \mu_{no\ cancel}$
- `ttest_ind (ë…ë¦½ í‘œë³¸ t-test)`ì— ëŒ€í•´ ì•Œì•„ë³´ì„¸ìš”.
  - `scipy` ê³µì‹ ë¬¸ì„œ : [ttest_ind](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.ttest_ind.html)
  - ê³µì‹ ë¬¸ì„œì˜ `parameter`ë“¤ì´ ì˜ë¯¸í•˜ëŠ” ê²ƒì— ì£¼ì˜í•´ì„œ ì‚´í´ë³´ì„¸ìš”.
    - ë“±ë¶„ì‚°ì„±ì— ëŒ€í•œ íŒŒë¼ë¯¸í„°ê°€ ë¬´ì—‡ì¸ì§€ í™•ì¸í•´ë³´ê³ , ì´ë¥¼ ì–´ë–»ê²Œ ì„¤ì •í•˜ë©´ ì¢‹ì„ì§€ ìƒê°í•´ë³´ì„¸ìš”.
- `adr` ì»¬ëŸ¼ ëŒ€ì‹  ë‹¤ë¥¸ ì»¬ëŸ¼ë“¤ë¡œë„ ìœ„ì˜ ê³¼ì •ì„ í†µí•´ ê°€ì„¤ì„ ìˆ˜ë¦½í•˜ê³  í†µê³„ì ìœ¼ë¡œ ê²€ì •í•´ë³´ì„¸ìš”! (optional)

````python
stats.ttest_ind(cancel_adr, no_cancel_adr, equal_var=False, alternative="two-sided")
````

ğŸ’¡ **í•´ì„¤ Tip)** <br>
> ë“±ë¶„ì‚°ì„±ì„ ë§Œì¡±í•˜ì§€ ì•Šê¸° ë•Œë¬¸ì—, `equal_var=False`ì„ ì§€ì •í•©ë‹ˆë‹¤.<br>

  - ì‹ ë¢°ìˆ˜ì¤€ 95%í•˜ì—, ì·¨ì†Œëœ ì˜ˆì•½ê³¼ ì·¨ì†Œë˜ì§€ ì•Šì€ ì˜ˆì•½ì˜ í‰ê·  `adr`ì€ ê°™ì§€ ì•Šë‹¤ëŠ” **ëŒ€ë¦½ê°€ì„¤ì€ í†µê³„ì ìœ¼ë¡œ ìœ ì˜í•©ë‹ˆë‹¤**.

  - [ì´í•­ë¶„í¬ reference](https://viewpds.jihak.co.kr/tsoldb/%EA%B5%90%EA%B3%BC%EC%84%9C%EB%B3%84/%EA%B3%A0_%EC%A0%81%EB%B6%84%EA%B3%BC%20%ED%86%B5%EA%B3%84(%EC%9D%B4%EA%B0%95%EC%84%AD)/4-1/%EC%A7%80%ED%95%99_%EC%A0%81%EB%B6%84%EA%B3%BC%20%ED%86%B5%EA%B3%84_%EA%B3%A0_%EC%9D%B4_4_1_%ED%99%95%EB%A5%A0%EB%B6%84%ED%8F%AC_%EA%B5%90%EA%B3%BC%EC%84%9C%20%EB%B3%B8%EB%AC%B8.pdf)
- ì•„ë˜ì˜ ì½”ë“œë¥¼ ì‹¤í–‰í•˜ì—¬, ì´í•­ë¶„í¬ê°€ ì–´ë–¤ ëª¨ì–‘ì„ ê°–ëŠ”ì§€ ì‚´í´ë³´ì„¸ìš”.


````python
# í™•ë¥ ë³€ìˆ˜ Xê°€ ì´í•­ë¶„í¬ B(n, p)ë¥¼ ë”°ë¥¼ ë•Œ, nì´ ì¶©ë¶„íˆ í¬ë‹¤ë©´ XëŠ” ì •ê·œë¶„í¬ N(np, npq)ë¥¼ ë”°ë¥¸ë‹¤ (ë‹¨, q=1-p)
from math import factorial

# Probability density of the binomial distribution
# pì˜ í™•ë¥ ì„ ê°–ëŠ” níšŒì˜ ì‹œí–‰ì—ì„œ kë²ˆ ì„±ê³µí•  í™•ë¥ ì„ ë°˜í™˜í•˜ëŠ” bin_dist í•¨ìˆ˜ ìƒì„±
def bin_dist(k, n, p):
    nck = factorial(n) / (factorial(k) * factorial(n - k))
    pd = nck * p**k * (1-p)**(n-k)
    return pd

x = np.arange(101)

# 100íšŒ ì‹œí–‰í–ˆì„ ë•Œ, 1ì´ ë‚˜ì˜¬ í™•ë¥  
# (e.g. ì•ë©´ì´ ë‚˜ì˜¬ í™•ë¥ ì´ ab_hotel.is_canceled.mean()ì¸ ë™ì „ì„ 100ë²ˆ ë˜ì¡Œì„ ë•Œ, ì•ë©´ì´ ë‚˜ì˜¬ í™•ë¥ )
pd1 = np.array([bin_dist(k, (101-1), ab_hotel.is_canceled.mean()) for k in range(101)])
# yì¶• ë²”ìœ„ ì§€ì •
plt.ylim(0, 0.1)

# ê·¸ë˜í”„ìƒì˜ x, yìœ„ì¹˜ì— ë¬¸ìì—´ ë‚˜íƒ€ë‚´ëŠ” plt.text
plt.text(80, 0.095, f'n, p = {101-1}, {round(ab_hotel.is_canceled.mean(), 2)}')
plt.bar(x, pd1, color='lightcoral')
plt.title("ì˜ˆì•½ ì·¨ì†Œìœ¨ 100ê±´ì˜ ì´í•­ ë¶„í¬", fontsize=13, fontweight="bold")
plt.show()
````

#### **3-4. `z-test`**
- `z-test`ë¥¼ í™œìš©í•˜ì—¬, ê°€ì„¤ì„ ê²€ì •í•´ë´…ì‹œë‹¤.
- statsmodel ê³µì‹ ë¬¸ì„œ : [statsmodels.stats.proportion.proportions_ztest](https://www.statsmodels.org/stable/generated/statsmodels.stats.proportion.proportions_ztest.html)
  - ê³µì‹ ë¬¸ì„œì˜ `parameter`ë“¤ì´ ì˜ë¯¸í•˜ëŠ” ê²ƒì— ì£¼ì˜í•´ì„œ ì‚´í´ë³´ì„¸ìš”.


````python
import statsmodels.api as sm

# ëŒ€ì¡°êµ°
num_ctrl = ab_hotel.query('button == "top"').shape[0]
num_exp = ab_hotel.query('button == "bottom"').shape[0]

# ëŒ€ì¡°êµ°, ì‹¤í—˜êµ° ê·¸ë£¹ ë‚´ retentonì´ Trueì¸ ìœ ì €ìˆ˜
top_cancel = ab_hotel.query('button == "top"').is_canceled.sum()
bottom_cancel = ab_hotel.query('button == "bottom"').is_canceled.sum()

# Z-test
z_score, p_value = sm.stats.proportions_ztest([top_cancel, bottom_cancel], 
                                              [num_ctrl, num_exp], 
                                              alternative= 'larger')

print(f'p-value (Z-test): {p_value}')
````
