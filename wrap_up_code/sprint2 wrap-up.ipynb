$$\large MAE=\frac{1}{n}\sum_{i=1}^{n}|y_{i}-\hat{y}_{i}|$$

$$\large MSE=\frac{1}{n}\sum_{i=1}^{n}(y_{i}-\hat{y}_{i})^2$$

$$\large RMSE=\sqrt{\frac{1}{n}\sum_{i=1}^{n}(y_{i}-\hat{y}_{i})^2}$$

$$\large R^2=\frac{SSR}{SST}=1-\frac{SSE}{SST}=1-\frac{\sum_{i=1}^{n}(y_{i}-\hat{y}_{i})^2}{\sum_{i=1}^{n}(y_{i}-\bar{y}_{i})^2}$$

```python
# 성능 평가
print('MAE:', mean_absolute_error(y_test, y_pred))
print('MSE:', mean_squared_error(y_test, y_pred))
print('RMSE:', mean_squared_error(y_test, y_pred)**(1/2))
print('R2-Score:', r2_score(y_test, y_pred))
print('Score:', model.score(?, ?))
```

# **B.분류 모델**

- Decision Tree 알고리즘으로 분류 모델을 만듭니다.

### **1.데이터 준비**

- 대상 데이터를 가져와 이후 과정을 준비합니다.

```python
# 데이터 읽어오기
path = 'https://raw.githubusercontent.com/Jangrae/csv/master/admission_simple.csv'
data = pd.read_csv(path)

# 확인
data.head()
```

```python
# target 확인
target = 'ADMIT'

# 데이터 분리
X = data.drop(target, axis=1)
y = data.loc[:, target]

# 함수 불러오기
from sklearn.model_selection import train_test_split

# 7:3로 분리
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)  
```

### **2.모델링**

- 본격적으로 모델을 선언하고 학습하고 평가하는 과정을 진행합니다.
- 우선 회귀 문제인지 분류 문제인지 명확히 구분합니다.

```python
# 1단계: 불러오기
from sklearn.tree import DecisionTreeClassifier
```

```python
# 2단계: 선언하기
model = DecisionTreeClassifier()
```

```python
# 3단계: 학습하기
model.fit(X_train, y_train)
```

```python
# 4단계: 예측하기
y_pred = model.predict(X_test)
```

```python
# 실젯값, 예측값 비교
print(y_test.values[:10])
print(y_pred[:10])
```

### **3.성능 평가**

- 다양한 성능 지표로 분류 모델 성능을 평가합니다.

**1) Confusion Matrix**

```python
# 함수 불러오기
from sklearn.metrics import confusion_matrix

# 성능 평가
print(confusion_matrix(y_test, y_pred))
```

```python
# 혼동행렬 시각화
plt.figure(figsize=(5, 3))
sns.heatmap(confusion_matrix(y_test, y_pred), 
            annot=True, 
            cmap='Blues',
            cbar=False,
            xticklabels=['0(Fail)', '1{Pass)'],
            yticklabels=['0(Fail)', '1{Pass)'],
            annot_kws={'size': 12})
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

**2) Accuracy, Precision, Recall, F1**

$$\large Accuracy = \frac{TP+TN}{TP+TN+FP+FN}$$

$$\large Precision = \frac{TP}{TP+FP}$$

$$\large Recall = \frac{TP}{TP+FN}$$

$$\large F1 = \frac{2\times Precision\times Recall}{Precision+Recall}$$

```python
# 함수 불러오기
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# 성능 평가
print('Accuracy:', accuracy_score(y_test, y_pred))
print('Score:', model.score(X_test, y_test))
print('Precision:', precision_score(y_test, y_pred, average=None))
print('Recall:', recall_score(y_test, y_pred, average=None))
print('F1-Score:', f1_score(y_test, y_pred, average=None))
```

**3) Classification Report**

```python
# 함수 불러오기
from sklearn.metrics import classification_report

# 성능 평가
print(classification_report(y_test, y_pred))
```

```python
# 변수 제거
del_cols = ['id']
data.drop(del_cols, axis=1, inplace=True)

# 결측치 제거
data.dropna(inplace=True)

# 확인
data.head()
```
